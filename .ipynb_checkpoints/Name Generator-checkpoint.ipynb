```python
import re
import random
import torch.nn as nn
import torch.optim as optim
import pickle as p
from model import NameModel
from utils import *
```

```python
file = open('movie_mystery.txt', 'r', encoding='utf-8')
text_origin = file.read()
file.close()
```

```python
l1 = re.findall(r'alt=".{0,50}"\sx', text_origin)
movie_mystery = [re.findall(r'=".{0,50}"',i)[0][2:-1] for i in l1]
print(movie_mystery)
```

```python
file = open("movie_mystery_list.txt", "w", encoding='utf-8')
for i in movie_mystery:
    print(i, file=file)
file.close()
```

```python
word_to_ix = {}
for sent in movie_mystery:
    for word in sent:
        if word not in word_to_ix:
            word_to_ix[word] = len(word_to_ix)
word_to_ix['<EOP>'] = len(word_to_ix)
word_to_ix['<START>'] = len(word_to_ix)

VOCAB_SIZE = len(word_to_ix)
```

```python
file = open('wordDic_movie_mystery', 'wb')
p.dump(word_to_ix, file=file)
file.close()
```

```python
data = movie_mystery.copy()
for i in range(len(movie_mystery)):
    data[i] = ['<START>'] + toList(movie_mystery[i]) + ['<EOP>']
print(data)
```

```python
model = NameModel(len(word_to_ix), 256, 256)
model.cuda()  # running on GPU,if you want to run it on CPU,delete all .cuda() usage.
optimizer = optim.RMSprop(model.parameters(), lr=0.01, weight_decay=0.0001)
criterion = nn.NLLLoss()
```

```python
one_hot_var_target = {}
for w in word_to_ix:
    one_hot_var_target.setdefault(w, make_one_hot_vec_target(w, word_to_ix))
```

```python
epochNum = 10
TRAINSIZE = len(data)
batch = 100
def test():
    v = int(TRAINSIZE / batch)
    loss = 0
    counts = 0
    for case in range(v * batch, min((v + 1) * batch, TRAINSIZE)):
        s = data[case]
        hidden = model.initHidden()
        t, o = makeForOneCase(s, one_hot_var_target)
        output, hidden = model(t.cuda(), hidden)
        loss += criterion(output, o.cuda())
        counts += 1
    loss = loss / counts
    print("=====",loss.data[0])
print("start training")
for epoch in range(epochNum):
    for batchIndex in range(int(TRAINSIZE / batch)):
        model.zero_grad()
        loss = 0
        counts = 0
        for case in range(batchIndex * batch, min((batchIndex + 1) * batch, TRAINSIZE)):
            s = data[case]
            hidden = model.initHidden()
            t, o = makeForOneCase(s, one_hot_var_target)
            output, hidden = model(t.cuda(), hidden)
            loss += criterion(output, o.cuda())
            counts += 1
        loss = loss / counts
        loss.backward()
        print(epoch, loss.data[0])
        optimizer.step()
    test()
torch.save(model, 'name-gen.pt')
```
